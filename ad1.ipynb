```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
terror_data = pd.read_csv('dane.csv', encoding='ISO-8859-1',
                          usecols=[0, 1, 2, 3, 8, 11, 13, 14, 35, 84, 100, 103])
terror_data = terror_data.rename(
    columns={'eventid':'id', 'iyear':'year', 'imonth':'month', 'iday':'day',
             'country_txt':'country', 'provstate':'state', 'targtype1_txt':'target',
             'weaptype1_txt':'weapon', 'nkill':'fatalities', 'nwound':'injuries'})
```

```python
# Ile było przestępstw w danym kraju?
plt.subplots(figsize=(15,6))
sns.countplot('country',data=terror_data,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Liczba przestępstw w danych krajach')
plt.show()
```

```python
sns.set(font_scale=1.5)
fig = plt.figure(figsize=(20,10))
sns.countplot('day',data=terror_data);
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.title('Terrorist attacks', fontsize=20)
plt.tight_layout()
```
